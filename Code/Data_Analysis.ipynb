```python
import glob
```

```python
weight_list = ['yolov5s_best', 'yolov5s_last', 
               'yolov5m_best', 'yolov5m_last',
               'yolov5l_best', 'yolov5l_last',
               'yolov5x_best', 'yolov5x_last']
```

```python
prc_time = [7.293233394622803,
7.498827219009399,
9.75796103477478,
9.768436193466187,
13.286372661590576,
13.517300605773926,
17.506194353103638,
17.820310831069946]
```

```python
file_list = glob.glob('C:/DEV/STUDY/maskrcnn/yolov5/runs/yolo_detect/yolov5x_last/*.jpg')

df = pd.DataFrame()

for n in file_list:
    file_name = n[(n.find('\\')+1):]
    Y = file_name[:file_name.find('-')]
    df_temp = pd.DataFrame({'f_name' : [file_name], 'Y' : [Y]})
    df = pd.concat([df, df_temp])

df = df.iloc[5:]

df
```

```python
df['predY'] = ['block'] * 16 + ['inner'] * 18 + ['outer'] * 4 + ['ok'] * 75 + ['outer'] * 19 + ['scar'] * 15 + ['outer'] * 2 + ['tear'] * 8 + ['outer'] * 7
df
```

```python
df_total = pd.DataFrame()
```

```python
# 정확도
acc_rate = sum(df['Y'] == df['predY']) / len(df)
# 1종 오류
type1_err = sum(df[df['Y'] == 'ok']['predY'] != 'ok')
# 2종 오류
type2_err = sum(df[df['Y'] != 'ok']['predY'] == 'ok')
# 불량 카테고리 분류 정확도
err_acc = (sum(df[df['Y'] != 'ok']['Y'] == df[df['Y'] != 'ok']['predY'])) / (sum(df['Y'] != 'ok'))

df_temp = pd.DataFrame({'m_name' : ['yolov5x_last'], 'ACC' : [acc_rate], 'Time' : [prc_time[7]], 
                        'type1_err' : [type1_err], 'type2_err' : [type2_err], 'err_acc' : [err_acc]})

df_total = pd.concat([df_total, df_temp])
df_total
```

```python
df_total.reset_index(inplace=True, drop=True)
df_total
```

```python
df_total.set_index('m_name', inplace=True)
df_total
```
