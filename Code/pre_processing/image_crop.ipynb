```python
from PIL import Image, ImageChops
import glob
import os
```

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        pass
```

```python
def crop(dir_path):
    images = glob.glob(f'{dir_path}/*.jpg')
    for i in range(len(images)):
        im = Image.open(images[i])
        background = Image.new(im.mode, im.size, im.getpixel((0, 0)))
        diff = ImageChops.difference(im, background)
        diff = ImageChops.add(diff, diff, 2.0, -35)
        bbox = diff.getbbox()
        bbox_lst = list(bbox)
        bbox_lst[0] -= 5
        bbox_lst[1] -= 5
        bbox_lst[2] += 5
        bbox_lst[3] += 5
        if bbox:
            createFolder(f'{dir_path}/crop')
            im.crop(bbox_lst).resize((128,128)).save(f'{dir_path}/crop/{i}.jpg', 'jpeg')
        else:
            print('Failure!')
            return
```

```python
# 테스트
dir_path = 'C:/ASH/intown_project/Data/Oring_cut/standard/OK'
crop(dir_path)
```

```python
# 정상 이미지 리사이즈
image_type = ['standard', 'Max_Exposure', 'Flash']
for i in image_type:
    dir_path = f'C:/ASH/intown_project/Data/Oring_cut/{i}/OK'
    crop(dir_path)
```

```python
# 불량 이미지 리사이즈
ng_dir = ['block', 'inner', 'outer', 'scar', 'tear']
image_type = ['standard', 'Max_Exposure', 'Flash']
for i in image_type:
    for j in ng_dir:
        dir_path = f'C:/ASH/intown_project/Data/Oring_cut/{i}/NG/{j}'
        crop(dir_path)
```
